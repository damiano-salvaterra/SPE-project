```python
import numpy as np
import matplotlib.pyplot as plt
from ChannelModel import *


model = ChannelModel(
    freq=2.4e9,        # 2.4 GHz typical Wi-Fi frequency
    coh_d=10,          # coherence distance in meters
    shadow_dev=9,    # shadowing std deviation in dB
    dspace_step=1,     # 1 meter grid step
    dspace_npt=100     # 100x100 spatial grid
)

# Genera la shadowing map
shadowing_map = model.generate_shadowing_map()

# Plot della mappa
plt.figure(figsize=(8, 6))
plt.contourf(model.X, model.Y, shadowing_map, levels=100, cmap='viridis')
plt.colorbar(label='Shadowing (dB)')
plt.title("Generated Shadowing Map (Gudmundson Model)")
plt.xlabel("x [m]")
plt.ylabel("y [m]")
plt.axis("equal")
plt.tight_layout()
plt.show()

# Statistiche di controllo
print(f"Media = {np.mean(shadowing_map):.2f} dB")
print(f"Deviazione standard = {np.std(shadowing_map):.2f} dB (atteso: {model.shadow_dev} dB)")
```
